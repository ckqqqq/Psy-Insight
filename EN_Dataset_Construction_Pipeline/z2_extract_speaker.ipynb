```python
def replace_list(orgin_str,replace_list:dict):
    tmp_str=orgin_str# 字符串直接就是值的复制
    for old_str,new_str in replace_list.items():
        tmp_str=tmp_str.replace(old_str,new_str)
    return tmp_str
```

```python
def add_text_participant_num(old_data):
    new_data=[]
    unit_id=0
    participant_list=[]
    dialog_content_replace_dict={")":"]","(":"["}
    for x,item in enumerate(old_data):
        if item["key"]=="fewshot":
            parti_set=set()
            for x,dialog_item in enumerate(item["dialog"]):
                orgin_text=dialog_item["origin_text"]
                name=orgin_text.split(":")[0]
                parti_set.add(name)
                item["dialog"][x]["id"]=str(unit_id)+"-"+str(x)
                dialog_text=orgin_text[len(name)+1:]
                dialog_text=replace_list(dialog_text,dialog_content_replace_dict)
                labels_list=[]
                if "action_labels" in dialog_item:
                    for sub_str in dialog_item["action_labels"]:
                        sub_str=replace_list(sub_str,dialog_content_replace_dict)
                        beg_idx=dialog_text.find(sub_str)
                        dialog_text=dialog_text[:beg_idx]+dialog_text[beg_idx+len(sub_str):]
                        labels_list.append({
                        "label": sub_str,
                        "index": beg_idx,
                        "type": ""})
                if len(labels_list)!=0:
                    item["dialog"][x]["labels"]=labels_list
                if "action_labels" in dialog_item:
                    item["dialog"][x].pop("action_labels")
                item["dialog"][x]["participant"]=name
                if dialog_text==" " and labels_list:
                    dialog_text=labels_list[0]["label"]
                item["dialog"][x]["text"]=dialog_text
                item["dialog"][x].pop("origin_text")
                

                
            participant_list=sorted(list(parti_set))
            item["unit_id"]=unit_id
            item["participant_num"]=len(participant_list)
            item["participant"]=participant_list
            item.pop("Participant")
            
            unit_id+=1
        new_item=item
        new_data.append(new_item)
    return new_data
proc_en_data=add_text_participant_num(en_data)
```
