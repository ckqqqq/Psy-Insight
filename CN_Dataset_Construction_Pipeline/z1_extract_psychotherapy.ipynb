```python
""" 
json.dumps(): 将 Python 对象编码成 JSON 字符串。 json.dump(): 将 Python 对象编码成 JSON 字符串，并写入到文件中。json.loads(): 将已编码的 JSON 字符串解码为 Python 对象。json.load(): 读取文件中的 JSON 编码的数据，并解码为 Python 对象。
"""
import os
import json
import torch
torch.cuda.is_available()
def json_standard(input_object):

    # print(input_object,type(input_object))
    return json.dumps(input_object,ensure_ascii=False,indent=4)
def read_file(file_path:str):
    with open(file_path,"r",encoding="utf-8") as i_f:
        return json.load(i_f)
def write_file(file_path:str,obj,overwrite=False):
    if os.path.exists(file_path)==False or overwrite==True:
        with open(file_path,"w",encoding="utf-8") as o_f:
            json.dump(obj,o_f,ensure_ascii=False)
        print(f"写入{file_path}")
    else:
        print(f"{file_path}文件已经存在")
def check_file(file_path:str):
    """返回该路径下是否存在文件"""
    is_exist=os.path.exists(file_path)
    if is_exist:
        file_name=file_path.split("/")[-1]
        print(f"{file_name} exist")
    return is_exist

import torch
import json

from transformers import (AutoTokenizer)
model_id="/home/ckqsudo/code2024/0model/Mistral-7B-Instruct-v0.2"
tokenizer=AutoTokenizer.from_pretrained(model_id)
# 这里不设置 trust_remote_code也行
tokenizer.padding_size='right'#填充符的位置
```

```python
import copy

def init_topic_dialog():
    topic_dialog_unit={
                      'theme':'',
                      'chat_stage':'',
                      'is_same_qa':'',
                      'is_same_case':'',
                      'background':'',
                      'pre_reasoning': '',
                      'topic_dialog':None,
                      'post_reasoning':'',
                      'extra_tags':[]}
    return topic_dialog_unit.copy()

def format_dialog_item(item):
    local_item={
        "key": "fewshot",
        "unit_id": item["unit_id"],
        "method": item["method"],
        "dialog": []
    }
    for i in item['dialog']:
        if "speaker" in i:
            del i['speaker']
        if 'labels' in i:
            for x,ii in enumerate(i['labels']):
                if i['labels'][x]['type']=='Unknown':
                    i['labels'][x]['type']=label_dict[i['labels'][x]['label']]

                if i['labels'][x]['type']=='sentence_reasoning':
                    sr=i['labels'][x]["label"]
                    if "reasoning" in i:
                        i["reasoning"]+=" "+sr[1:-1]
                    else:
                        i["reasoning"]=sr[1:-1]

                elif i['labels'][x]['type'].endswith("_act") and i["text"][0]!="（" and i['labels'][x]["label"] not in ['（来访者暧昧对象）']:
                    i["text"]=i['labels'][x]["label"]+i["text"]

        local_item["dialog"].append(i)
    return local_item.copy()


def div_topic_dialog(cur_data):
    result=[]
    cur_chat_stage=""
    cur_topic=init_topic_dialog()
    for x,item in enumerate(cur_data):
        if item['key']=='dialog_id':
            if 'type' in item:
                cur_topic['extra_tags'].append({"type":item['type']})
            if cur_topic['topic_dialog']!=None:
                result.append(cur_topic)
                cur_topic=init_topic_dialog()
        elif item['key']=='chat_stage':
            cur_topic['chat_stage']=item['content']

        elif item['key']=='reasoning' and cur_topic['topic_dialog']==None:
            cur_topic['pre_reasoning']+=item['content']
        elif item['key']=='reasoning' and cur_topic['topic_dialog']!=None:
            cur_topic['post_reasoning']+=item['content']
        elif item['key']=='fewshot':
            # print("fewshot！")
            
            if cur_topic['topic_dialog']==None:
                cur_topic['topic_dialog']=format_dialog_item(item)
            else:# 处理
                result.append(copy.deepcopy(cur_topic))
                cur_topic=init_topic_dialog()
                cur_topic['topic_dialog']=format_dialog_item(item)
                cur_topic['pre_reasoning']=result[-1]['post_reasoning']
        else:
            print(item)
    return result
cur_data=copy.deepcopy(cn_data)
result=div_topic_dialog(cur_data)
```
