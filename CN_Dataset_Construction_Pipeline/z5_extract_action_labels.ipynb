```python
quote_list = [
    # 中文引号
    "“",
    "“",
    "”",
    "‘",
    "’",
    # 英文引号
    "〝",
    "〞",
    "‘",
    "’",
    # 其他类型的引号
    "『",
    "』",
    "｢",
    "｣",
    "《",
    "》",
    # 特殊符号
    # 中文引号（全角）
    "「",
    "」",
    # 中文引号（半角）
    "『",
    "』",
    # 英文引号（全角）
    "﹁",
    "﹂",
    # 英文引号（半角）
    "﹃",
    "﹄",
    # 日文引号
    "「",
    "」",
    # 韩文引호
    "「",
    "」",
]
def del_comma(old_data,new_data):
    if len(new_data)>0:
        return
    dialog_unit_id=0
    for idx,item in enumerate(old_data):
        if item['key']=="fewshot":
            item['unit_id']=dialog_unit_id
            dialog_unit_id+=1
            for y,dialog_item in enumerate(item['dialog']):
                d_text=dialog_item['text']
                # quote_list
                if d_text!="" and d_text[0] in quote_list:
                    d_text=d_text[1:]
                    while d_text!="" and d_text[0] in quote_list:
                        d_text=d_text[1:]
                if d_text=="":
                    del item['dialog'][y]
                    continue 
                if "labels" in dialog_item and dialog_item['text'][0] in quote_list:
                    for z,dic in enumerate(dialog_item["labels"]):
                        pre_index=dic["index"]
                        if pre_index>0:
                            item["dialog"][y]["labels"][z]["index"]=pre_index-1
                item["dialog"][y]['text']=d_text
        new_data.append(item)
```

```python

del_comma_data=[]
del_comma(add_participant_data,del_comma_data)
```

```python
def add_dialog_item_id(old_data,new_data):
    if len(new_data)!=0:
        print("!")
        return
    for x,item in enumerate(old_data):
        if item["key"]=="fewshot":
            topic_id=item['unit_id']
            for y,dialog_item in enumerate(item['dialog']):
                d_text=dialog_item['text']
                while d_text!="" and d_text[0] in ["“","\""]:
                    d_text=d_text[1:]
                while d_text!="" and d_text[-1] in ["”","\""]:
                    d_text=d_text[:-1]
                item['dialog'][y]['text']=d_text
                item['dialog'][y]['id']=str(topic_id)+"-"+str(y)
        new_data.append(item)
```

```python
add_new_id_data=[]
add_dialog_item_id(del_comma_data,add_new_id_data)
```
