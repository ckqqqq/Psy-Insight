```python
import json
input_file_path="/home/ckqsudo/code2024/0dataset/E-bed/English/z1_datastream_划分reasoning和对话/z2划分后的json文件/English_Dataset.json"
with open(input_file_path,"r",encoding="utf-8") as i_f:
    en_data=json.load(i_f)
```

```python
count_dialog=0
count_turn=0
for item in en_data:
    if item['key']=='fewshot':
        print(len(item['dialog']))
        count_dialog+=1
        count_turn+=len(item['dialog'])
```
